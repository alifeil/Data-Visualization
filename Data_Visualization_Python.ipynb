```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
##print(t1) untuk menampilkan data yang ada di t1

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
## t2, f(t2), 'k' kalo gapake mereka cuman 'bo' berarti nanti gak ada garisnya

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
##contoh plot ketiga dibawah ini cuman bisa dua
##plt.subplot(212)
##plt.plot(t2, np.cos(2*np.pi*t2),'g--')
##plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

# Fixing random state for reproducibility
np.random.seed(19680801)

N_points = 100000
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins, color="yellow", alpha = 0.5)
axs[1].hist(y, bins=n_bins)
```

```python
fig, axs = plt.subplots(1, 2, tight_layout=True)

# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = axs[0].hist(x, bins=n_bins)

# We'll color code by height, but you could use any scalar
fracs = N / N.max()

# we need to normalize the data to 0..1 for the full range of the colormap
norm = colors.Normalize(fracs.min(), fracs.max())

# Now, we'll loop through our objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))##cm itu palet warna kayak viridis
    ##contoh seperti comment dibawah ini
    #color = plt.cm.inferno(norm(thisfrac))
    #color = plt.cm.magma(norm(thisfrac))
    thispatch.set_facecolor(color)

# We can also normalize our inputs by the total number of counts
axs[1].hist(x, bins=n_bins, density=True)

# Now we format the y-axis to display percentage
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))
```

```python
fig, axs = plt.subplots(3, 1, figsize=(5, 15), sharex=True,
sharey=True,
                        tight_layout=True)
# We can increase the number of bins on each axis
axs[0].hist2d(x, y, bins=40)
# As well as define normalization of the colors
axs[1].hist2d(x, y, bins=40, norm=colors.LogNorm())
# We can also define custom numbers of bins for each axis
axs[2].hist2d(x, y, bins=(80, 10), norm=colors.LogNorm())
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# data to plot
n_groups = 4
means_frank = (90, 55, 40, 65) ##data biru
means_guido = (85, 62, 54, 20) ##data hijau

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35 #untuk mengatur lebar dari bar
opacity = 0.8

rects1 = plt.bar(index, means_frank,
bar_width, alpha=opacity,color='b',
label='Frank')

rects2 = plt.bar(index + bar_width,
means_guido, bar_width,
alpha=opacity,
color='g',
label='Guido')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('A',
'B', 'C', 'D')) ##bar_width untuk menempatkan posisi di tengah - tengah
plt.legend()

plt.tight_layout()
```
